```python
import numpy as np
import pandas as pd
```

# LOAD DATA

```python
data_preprocessed = pd.read_csv(r"C:\Users\Bhaypreet\OneDrive\Pictures\IIT ROPAR\data_science_projects\Absenteeism\Absenteeism_Preprocess.csv")
```

```python
data_preprocessed.head()
```

# Creating targets ; Moderately and Highly absent

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] > data_preprocessed['Absenteeism Time in Hours'].median(),1 ,0)
```

```python
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
```

```python
data_preprocessed.head()
```

# Balancing Targets

```python
targets.sum()/targets.shape[0] 
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours'],axis=1)
```

```python
data_with_targets.head()
```

# Input data

```python
data_with_targets.shape
```

```python
input_data = data_with_targets.iloc[:,:-1]  
```

# STANDARDIZING DATA

```python
from sklearn.preprocessing import StandardScaler
absenteeism_scaler = StandardScaler()
```

```python
absenteeism_scaler.fit(input_data)
```

```python
standardized_inputs = absenteeism_scaler.transform(input_data) 
```

```python
standardized_inputs.shape
```

```python
input_data.columns.values
```

# Train_Test_Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test ,y_train ,y_test = train_test_split(standardized_inputs,targets , train_size = 0.8 , random_state = 20)
```

```python
print(x_train.shape ,x_test.shape ,y_train.shape ,y_test.shape)
```

# TRAINING THE MODEL : Logistic_Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
reg = LogisticRegression()
reg.fit(x_train,y_train)
```

```python
x_pred = reg.predict(x_train)
```

```python
print("Training_Accuracy :", accuracy_score(x_pred,y_train))
```

# Manually checking the accuracy

```python
summary_table.index = summary_table.index +1
```

```python
summary_table = summary_table.sort_index()
```

```python
summary_table
```

# Interpreting the coef_

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Weights)
summary_table
```

```python
summary_table.sort_values('Odds_ratio',ascending = False)
```

# Testing the model

```python
predicted_proba = reg.predict_proba(x_test) 
predicted_proba
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Confusion Matrix:", confusion_matrix(y_test, y_pred))
```
