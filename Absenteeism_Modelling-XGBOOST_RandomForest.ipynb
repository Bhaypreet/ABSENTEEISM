# Creating target by making classes of moderately and highly absenteeism

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] > data_preprocessed['Absenteeism Time in Hours'].median(),1 ,0)
```

```python
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
```

# Balancing test and targets set

```python
targets.sum()/targets.shape[0]
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours'],axis=1)
```

```python
data_with_targets.head()
```

# Input Data

```python
data_with_targets.shape
```

```python
input_data = data_with_targets.iloc[:,:-1]  
```

# TRAIN_TEST_SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test ,y_train ,y_test = train_test_split(input_data,targets , train_size = 0.8 , random_state = 20) 
```

```python
print(x_train.shape ,x_test.shape ,y_train.shape ,y_test.shape)
```

# TRAINING/TESTING THE MODEL

```python
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
```

## XGBoost_Classifier

```python
XG = XGBClassifier(
    random_state=42,
    eval_metric='logloss'
)
```

```python
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [3, 5],
    'learning_rate': [0.01, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}
```

```python
model1 = GridSearchCV(
    estimator=XG,
    param_grid=param_grid,
    cv=5,  
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)
```

```python
model1.fit(x_train, y_train)
```

```python
x_pred1 = model1.predict(x_train)
print('Accuracy Score of Training :',accuracy_score(x_pred1,y_train))
```

```python
y_pred1 = model1.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred1))
print( classification_report(y_test, y_pred1))
print("Confusion Matrix:", confusion_matrix(y_test, y_pred1))
```

## RandomForest_Classifier

```python
rf = RandomForestClassifier(random_state=42)
```

```python
param_grid2 = {
    'n_estimators': [50, 1500, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'max_features': ['sqrt', 'log2']
}
```

```python
model2 = GridSearchCV(
    estimator=rf,
    param_grid=param_grid2,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)
```

```python
model2.fit(x_train, y_train)
```

```python
x_pred2 = model2.predict(x_train)
```

```python
print("Training_Accuracy : ",accuracy_score(x_pred2,y_train))
```

```python
y_pred2 = model2.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred2))
print( classification_report(y_test, y_pred2))
print("Confusion Matrix:", confusion_matrix(y_test, y_pred2))
```
